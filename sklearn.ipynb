```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
fulldata = pd.read_csv("data/train.csv")
data_test = pd.read_csv("data/test.csv")
fulldata.shape, data_test.shape
```

```python
fulldata.head()
```

```python
fulldata[fulldata.isnull().any(axis=1)]
```

```python
labels = fulldata['rainfall']
data = fulldata.drop(['rainfall'], axis=1)
labels.shape, data.shape
```

```python
test_data = pd.read_csv("data/test.csv")
test_data
```

```python
test_data[test_data.isnull().any(axis=1)]
```

```python
test_data = test_data.fillna(test_data.mean())
test_data[test_data.isnull().any(axis=1)]
```

```python
data = data.drop(['id'], axis=1)
test_data = test_data.drop(['id'],axis=1)
```

```python
data, test_data
```

```python
param_grid = {
    'n_estimators' : [ 50, 100, 200],
    'max_depth':[None, 5, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
```

```python
model = RandomForestRegressor()
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=5,
    scoring='neg_mean_absolute_error',
    n_jobs=-1
)
```

```python
grid_search.fit(data, labels)
```

```python
best = grid_search.best_estimator_
```

```python
preds_best = best.predict(test_data)
preds_best[:10]
```

```python
zxc = pd.read_csv("data/sample_submission.csv")
zxc.head()
```

```python
zxc['rainfall'] = preds_best
zxc.head()
```

```python
zxc.to_csv("data/zxcsk3.csv", index=False)
```
